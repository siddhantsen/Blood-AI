```python
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(40, 40))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
sum_abs_values = corr_matrix.abs().sum()
sum_abs_values.sort_values(ascending=False, inplace = True)
# Step 4: Plot the results using a bar plot
sum_abs_values.plot(kind='bar')
plt.ylabel('Sum of Absolute Values')
plt.title('Sum of Absolute Values for Each Column')
plt.fig_size = (10, 40)
plt.show()
```
